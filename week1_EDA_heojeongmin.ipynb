# 2. 유의미한 Feature 생성하기 (10개 이상)
- 2.1. 경매 소요일자
- 2.2. 초고층 빌딩의 여부
- 2.3. 로열층의 여부
- 2.4. 건물에서의 높이 비율
- 2.5. 실제 경매 건물 면적
- 2.6. 실제 경매 토지 면적
- 2.7. 재건축 프리미엄
- 2.8. 시별 평균가격 차이(=시세차익 가능성)
- 2.9. 감정가보다 낙찰가가 높은 프리미엄
- 2.10. 매물의 인기도

---

```python
## 2.1. 경매 소요일자
'''경매가 일자가 짧다면, 즉 경매가 빠르게 진행되었다면 인기있는 매물이라고 판단 할 수 있을 것입니다. 
따라서 최종경매일에서 최초경매일자를 뺀 경매 소요일자를 새로운 feature로 사용하도록 하겠습니다. ''' 
df['auction_date']  = df['Final_auction_date'].astype('datetime64[ns]') - df['First_auction_date'].astype('datetime64[ns]')
df['auction_date'].head()
```

```python
# 2.2. 초고층 빌딩의 여부
'''초고층 빌딩일 수록 가격에 대한 프리미엄이 붙어서 같은 지역 같은 평수라도 높은 가격이 형성될 것이라고 생각하였습니다. 
현재 기준에 따르면 초고층 빌딩은 총 층수가 50층 이상일때 적용됩니다. 
따라서 skyscraper라는 이름으로 변수를 추가하여 해당매물의 총 층수가 50층이상인 경우 1
그렇지 않은 경우 0으로 추가하도록 하였습니다. ''' 
df[df['Total_floor']>=50]
df["skyscraper"] = df["Total_floor"].apply(lambda x: 1 if x >= 50 else 0)
```

```python
# 2.3. 로열층 여부
'''
일반적으로 30층이상 일경우 로열층 프리미엄을 얻을 수 있습니다. 
따라서 해당 경매의 층수가 30층 이상일 경우 Royal_latyer라는 이름으로 변수를 추가하여 1의 값을 추가하고 그렇지 않을 경우 0으로 추가하도록 하였습니다.
''' 
df[df['Current_floor']>=30].head()
df["Royal_layer"] = df["Current_floor"].apply(lambda x: 1 if x >= 30 else 0)
```

```python
# 2.4 floor rate = 해당층/전체층
"""
경매 건물별로 건물의 총 층수와 해당 층수가 다릅니다. 또한 해당 층수에 대한 정보를 통해서 어느정도 높이에 위치에있는지를 알 수 있지만,
상대적인 비교를 하기에는 어렵습니다. 또한 일반적으로 건물의 층수가 높을 수록 매물의 가격이 오르지만
30층 이상의 로열층 혹은 50층이상의 초고층 빌딩이 있는 지역이 아니라면 그지역내에서의 비교를 하기에는 힘들 것입니다. 
따라서 해당층/전체건물층수를 통하여 해당 매물의 층의 rate를 알 수 있다면 같은지역의 다른 매물과의 비교를 통해서 층에 대한 로열티에 대해서 좀 더 자세하게 판단 할 수 있을 것입니다.
따라서 floor_rate라는 이름으로 해당건물에서 높이 비율을 변수로 추가하였습니다.  """
df["floor_rate"] = df['Current_floor'] / df['Total_floor']
df.head()
```

```python
# 2.5 Building Auction rate = 총건물 경매면적/총건물면적
#정보 출처: https://investnote8.tistory.com/24
'''데이터에서 존재하는 매물의 대부분은 주상복합과 아파트와 같은 건물들 입니다.
따라서 총 건물의 면적과 실제 경매를 통해 받을 수 있는 면적이 같은지 확인하는 것은 중요합니다.
만약 그 면적이 서로 다르다면 경매에 참여하기전 혹은 가격에 대한 분석을 하기전에 외부적인 이유를 확인해야 할 것입니다. 
따라서 이러한 비율을 building_rate이라는 이름의 변수로 추가하고 확인하도록 하겠습니다. '''
df["building_rate"] = df['Total_building_auction_area']/ df['Total_building_area'] 
df["building_rate"].tail(10)
```

```python
# 2.6. Land Acution rate = 총토지 경매면적/ 총토지 실면적
#정보 출처: https://www.ulsanpress.net/news/articleView.html?idxno=8541
'''경매받은 토지와 실면적의 차이로 많은 문제가 발생되고 있습니다. 
심지어 이러한 이유로 법적 논쟁까지 가는 경우도 많이 발생되고 있습니다. 
따라서 이러한 변수를 추가하여 해당매물이 적정 경매면적이 적당한 가격에 경매되고 있는지 확인하기 위하여
land_rate라는 이름으로 비율을 확인하도록 하겠습니다. '''

df["land_rate"] = df['Total_land_auction_area']/ df['Total_land_gross_area'] 
df["land_rate"].head(10)
```

```python
# 2.7 재건축 프리미엄
#정보 출처: ttps://www.yna.co.kr/view/AKR20150521000500003
'''
Preserve_regist_date, 즉 보존등기란 사람으로 따지면 출생신고와 비슷한 개념입니다.
즉, 건물의 소유주의 변경과 무관하게 해당 건물이 완공되어지고 주인이 바뀐 과정에 대해서 전체적으로 작성된 내용이라고 할 수 있습니다. 
또한 출처에 따르면 재건축을 위한 가능연수는 30년 이상이라고 나와있습니다. 따라서 이러한 정보들을 혼합하여
만약 건물의 연수가 30년이 넘었을 경우 재건축에 대한 기대심리를 반영한 프리미엄이 있을수 있다고 판단하여 Reconstruction라는 이름으로 재건축 가능여부를 변수로 추가하도록 하였습니다. '''


df['Reconstruction'] = 2020 - df["Preserve_regist_date"].str.slice(start=0, stop=4).astype('int')
df['Reconstruction'] = df["Reconstruction"].apply(lambda x: 1 if x >= 30 else 0)
df['Reconstruction'].tail(20)
```

```python
# 2.8 시별 평균 가격과의 차이 
#정보 출처: http://raemian.co.kr/community/times/realty/view.do?seq=1737&searchType=&searchStr=&pg=1
'''
위에서 그래프로 확인한 것처럼 도시에 따라서 그 평균가격이 차이가 존재하였습니다. 
이번에는 해당 도시안에서 해당 매물이 해당도시의 평균 가격보다 얼마나 차이나는지를 확인하여
시세차익의 가능성을 판단해 보도록 하겠습니다. '''

df['dong_mean'] = df.groupby(['addr_dong']).Hammer_price.transform('mean') #지역별 평균을 넣고
df['capital_gains'] = df['Hammer_price'] - df['dong_mean']
df[df['addr_dong']=="논현동"]
```

```python
# 2.9 감정가보다 낙찰가가 높은 프리미엄
#정보 출처: https://biz.chosun.com/site/data/html_dir/2011/03/13/2011031301410.html
'''
일반적으로 부동산 경매에 참여하는 사람들의 심리는 시세보다 저렴한 가격에 매물을 얻고자 하는 경우가 많습니다.
따라서 일반적으로 감정가보다 낙찰가가 낮게 책정되는 경우가 많습니다. 
하지만 시장 호황에 대한 분위기나 혹은 가격 상승에 대한 기대요인등 다양한 이유로 낙찰가가 감정가보다 높은 경매가 생기고 있습니다. 
따라서 감정가보다 낙찰가가 높은 경우 프리미엄에 대한 기대가 반영되었다고 판단하여
Expectation라는 이름으로 그 차이만큼을 변수로 추가하도록 하였습니다. '''

df['Expectation'] = df['Hammer_price' ] - df["Total_appraisal_price"]
df[df['Expectation']>0].head()
```

```python
# 2.10 매물의 인기도
#정보 출처: https://m.blog.naver.com/PostView.nhn?blogId=parks0626&logNo=140192996734&proxyReferer=https%3A%2F%2Fwww.google.com%2F
'''
경매에 나오는 부동산 매물은 유찰의 횟수가 많아질수록 약 10~20%정도 감가상각된 감정가를 가지고 다시 경매에 올라가게 됩니다. 
또한 데이터를 확인해본 결과 일반적으로 경매횟수는 유찰횟수보다 1번이 더 많게 형성되어있습니다. 대부분이 유찰이 0,경매가 1인 한번에 경매가 완료되었다고 판단할수있습니다.
하지만 유찰의 횟수가 훨씬 더 많은 경우도 있고 심지어 경매횟수가 많은 경우도 있었습니다.
전자의 경우는 단순히 매물의 인기가 없다고 판단할 수 있는 경우이고, 후자의 경우에는 낙찰받은 사람이 알수없는 이유로 다시 경매에 올린 경우라고 판단할 수 있습니다.
따라서 경매의 횟수 + 유찰의 횟수를 더하고 이에 역을 해주어 이것을 매물의 매력도라고 칭하였습니다. 
정상적인 매물은 이 점수가 1이지만 그렇지 않은 매물들은 점수가 0에 가까워 질 것입니다. '''
 
auctions_count = df['Auction_count'] - df['Auction_miscarriage_count']
auctions_count.sort_values().tail(10)

df['Attraction'] = 1/ (df['Auction_count'] - df['Auction_miscarriage_count'])
df['Attraction'].sort_values().head(10)
```
